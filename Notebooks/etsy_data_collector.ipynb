## Transform Data Into Relational Tables
- In this section, we took the 100+ columns (parameters) and used ChatGPT to strategic create database tables that group the right data together and relate it to the other tables through the Store ID
- The python code creates each table as an excel file
- We will update the code to create each table as a DB table once we scale


### Tables and Their Relationships

#### Stores Table
- Purpose: This table holds general information about each Etsy store.
- Columns:
- - Store ID (Primary Key)
- - Etsy Store Name
- - Store Details
- - Number of Admirers
- - Number of Overall Sales
- - Number of Items in Store
- - Number of Overall Reviews
- - Etsy Store Last Updated
- - Etsy Store URL
- Relationship: This table is the central entity, linked to all other tables through the Store ID.

#### Items Table
- Purpose: This table contains detailed information about each item sold by the stores.
- Columns:
- - Item ID (Primary Key)
- - Store ID (Foreign Key referencing Stores Table)
- - Item Name (e.g., Item Sold 1, Item Sold 2, etc.)
- - Title
- - Description
- - Price
- - Item Details
- - Ships From Location
- - Etsy POS Item Link
- - Reviews Count
- - Total Items Left in Inventory
- Relationship: Linked to the Stores Table through the Store ID. Each store can have multiple items, establishing a one-to-many relationship.

#### Social Media Table
- Purpose: This table captures the social media presence and activity of each store.
- Columns:
- - Social Media ID (Primary Key)
- - Store ID (Foreign Key referencing Stores Table)
- - Social Media Platform (e.g., Instagram, Facebook, Twitter)
- - Profile Name
- - Number of Followers
- - Number Following
- - Number of Overall Posts
- - Number of Posts in Last Month
- - Number of Posts in Last Week
- Relationship: Linked to the Stores Table through the Store ID. Each store can have multiple social media profiles, establishing another one-to-many relationship.

#### Meta Ads Table
- Purpose: This table tracks the Meta Ads data for each store.
- Columns:
- - Meta Ad ID (Primary Key)
- - Store ID (Foreign Key referencing Stores Table)
- - Ad Title
- - Ad Description
- - Ad Details
- - Advertiser Name
- - Posting Date
- - ChatGPT Visual Analysis
- - Video Transcript
- Relationship: Linked to the Stores Table through the Store ID. Each store can run multiple ads, creating a one-to-many relationship.

#### Engagement Metrics Table
- Purpose: This table contains engagement metrics such as shares, impressions, comments, and likes over time.
- Columns:
- - Engagement ID (Primary Key)
- - Store ID (Foreign Key referencing Stores Table)
- - Total Number of Items in Carts (Platform-Wide)
- - Total Number of Shares (Last 3 Months, 2 Months, 1 Month, 2 Weeks, 3 Days)
- - Total Number of Impressions (Last 3 Months, 2 Months, 1 Month, 2 Weeks, 3 Days)
- - Total Number of Comments (Last 3 Months, 2 Months, 1 Month, 2 Weeks, 3 Days)
- - Total Number of Likes (Last 3 Months, 2 Months, 1 Month, 2 Weeks, 3 Days)
- Relationship: Linked to the Stores Table through the Store ID. This table captures temporal engagement metrics for each store, establishing a one-to-one relationship with the Stores Table.

```python
# Creating individual DataFrames for each table and saving them to separate Excel files

# Stores Table
stores_columns = [
    "Store ID", "Etsy Store Name", "Store Details", "Number of Admirers", "Number of Overall Sales",
    "Number of Items in Store", "Number of Overall Reviews", "Etsy Store Last Updated", "Etsy Store URL"
]
stores_df = pd.DataFrame(columns=stores_columns)
stores_file_path = './Data Tables/Stores_Table.xlsx'
stores_df.to_excel(stores_file_path, index=False)

# Items Table
items_columns = [
    "Item ID", "Store ID", "Item Name", "Title", "Description", "Price", "Item Details", 
    "Ships From Location", "Etsy POS Item Link", "Reviews Count", "Total Items Left in Inventory"
]
items_df = pd.DataFrame(columns=items_columns)
items_file_path = './Data Tables/Items_Table.xlsx'
items_df.to_excel(items_file_path, index=False)

# Social Media Table
social_media_columns = [
    "Social Media ID", "Store ID", "Social Media Platform", "Profile Name", "Number of Followers", 
    "Number Following", "Number of Overall Posts", "Number of Posts in Last Month", "Number of Posts in Last Week"
]
social_media_df = pd.DataFrame(columns=social_media_columns)
social_media_file_path = './Data Tables/Social_Media_Table.xlsx'
social_media_df.to_excel(social_media_file_path, index=False)

# Meta Ads Table
meta_ads_columns = [
    "Meta Ad ID", "Store ID", "Ad Title", "Ad Description", "Ad Details", 
    "Advertiser Name", "Posting Date", "ChatGPT Visual Analysis", "Video Transcript"
]
meta_ads_df = pd.DataFrame(columns=meta_ads_columns)
meta_ads_file_path = './Data Tables/Meta_Ads_Table.xlsx'
meta_ads_df.to_excel(meta_ads_file_path, index=False)

# Engagement Metrics Table
engagement_metrics_columns = [
    "Engagement ID", "Store ID", "Total Number of Items in Carts (Platform-Wide)", 
    "Total Number of Shares (Last 3 Months)", "Total Number of Shares (Last 2 Months)", 
    "Total Number of Shares (Last 1 Month)", "Total Number of Shares (Last 2 Weeks)", 
    "Total Number of Shares (Last 3 Days)", "Total Number of Impressions (Last 3 Months)", 
    "Total Number of Impressions (Last 2 Months)", "Total Number of Impressions (Last 1 Month)", 
    "Total Number of Impressions (Last 2 Weeks)", "Total Number of Impressions (Last 3 Days)", 
    "Total Number of Comments (Last 3 Months)", "Total Number of Comments (Last 2 Months)", 
    "Total Number of Comments (Last 1 Month)", "Total Number of Comments (Last 2 Weeks)", 
    "Total Number of Comments (Last 3 Days)", "Total Number of Likes (Last 3 Months)", 
    "Total Number of Likes (Last 2 Months)", "Total Number of Likes (Last 1 Month)", 
    "Total Number of Likes (Last 2 Weeks)", "Total Number of Likes (Last 3 Days)"
]
engagement_metrics_df = pd.DataFrame(columns=engagement_metrics_columns)
engagement_metrics_file_path = './Data Tables/Engagement_Metrics_Table.xlsx'
engagement_metrics_df.to_excel(engagement_metrics_file_path, index=False)

(stores_file_path, items_file_path, social_media_file_path, meta_ads_file_path, engagement_metrics_file_path)
```
